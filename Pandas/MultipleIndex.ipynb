# Extract Rows From a MultiIndex DataFrame

```python
bigmac = pd.read_csv("bigmac.csv",parse_dates=["Date"],index_col = ["Date","Country"])
bigmac.sort_index(inplace = True)
bigmac.head()
```

```python
bigmac.loc[("2010-01-01", "Argentina"), "Price in US Dollars"]
```

```python
bigmac.ix[("2010-01-01", "Argentina"), 0] = 50
```

```python
bigmac.ix[("2010-01-01", "Argentina"), 0]
```

# .transpose() Method

```python
bigmac = pd.read_csv("bigmac.csv",parse_dates=["Date"],index_col = ["Date","Country"])
bigmac.sort_index(inplace = True)
bigmac.head()
```

```python
bigmac = bigmac.transpose()
bigmac.head()
```

```python
bigmac.ix["Price in US Dollars",("2016-01-01", "Argentina")]
```

```python
bigmac = pd.read_csv("bigmac.csv",parse_dates=["Date"],index_col = ["Country"])
bigmac.sort_index(inplace = True)
bigmac.head()
```

```python
bigmac.loc["Argentina"] = bigmac.loc["Argentina"].drop(bigmac.loc["Argentina"].columns[0], axis=1)
bigmac.head()
```

```python
bigmac.loc["Argentina"]
```

# .swaplevel()

```python
bigmac = pd.read_csv("bigmac.csv",parse_dates=["Date"],index_col = ["Date","Country"])
bigmac.sort_index(inplace = True)
bigmac.head()
```

```python
bigmac = bigmac.swaplevel()
bigmac.head(3)
```
