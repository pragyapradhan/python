## .apply()

```python
def convert_to_string_and_add_millions(number):
    return str(number) +"Million!"
bond["Box Office"] = bond["Box_Office"].apply(convert_to_string_and_add_millions)
```

```python
bond["Budget"] = bond["Budget"].apply(convert_to_string_and_add_millions)
```

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head()
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns:
    bond[col] = bond[col].apply(convert_to_string_and_add_millions)
bond.head(3)
```

```python
def d_r(row): 
    
    actor = row[1]
    budget = row[4]
    
    if actor == "Daniel Craig":
        return "The Best"
    elif actor == "Roger Moore" and budget >40:
        return "Enjoyable"
    else:
        return "I have no clue"
bond.apply(d_r, axis = 'columns')
```
