# .agg() Method

```python
ft = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = ft.groupby(["Sector","Industry"])
ft.head(3)
```

```python
sectors.agg({"Revenue" : "count",
             "Profits" : "sum",
            "Employees" : "mean"} )
```

```python
sectors.agg(["sum", "mean", "size"])
```

# Iterating through Groups

```python
ft = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = ft.groupby(["Sector","Industry"])
ft.head(3)
```

```python
sectors["Profits"].max()
```

```python
df = pd.DataFrame(columns = ft.columns)
df
```

```python
for sector,data in sectors:
    highest_revenue_company_ingroup = data.nlargest(5, "Revenue")
    df = df.append(highest_revenue_company_ingroup)
```

```python
df
```

```python
cities = ft.groupby("Location")
df = pd.DataFrame(columns = ft.columns)
df
```

```python
for city, data in cities:
    highest_revenue_in_city = data.nlargest(1,"Revenue")
    df = df.append(highest_revenue_in_city)
```
